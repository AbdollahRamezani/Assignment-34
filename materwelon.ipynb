```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
watermelon = cv2.imread("input/watermelon.jpg")
watermelon = cv2.cvtColor(watermelon, cv2.COLOR_BGR2RGB)
plt.imshow(watermelon)
```

```python
R, G, B = cv2.split(watermelon)
materwelon = cv2.merge([B, G, B])


plt.imshow(materwelon)
plt.title('Materwelon')
plt.axis("off")
plt.show()
```
