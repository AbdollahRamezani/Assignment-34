```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
r=np.zeros((300 , 500) ,dtype=np.uint8)
g=np.zeros((300 , 500) ,dtype=np.uint8)
b=np.zeros((300 , 500) ,dtype=np.uint8)

r[:,:]=255
g[:,:]=255
b[:,:]=255
```

```python
cv2.circle(r,(250,300), 230, 204, 25)
cv2.circle(g,(250,300), 230, 0, 25)
cv2.circle(b,(250,300), 230, 0,25)

cv2.circle(r,(250,300), 205, 255, 25)
cv2.circle(g,(250,300), 205, 153, 25)
cv2.circle(b,(250,300), 205,51 , 25)

cv2.circle(r,(250,300), 180, 255, 25)
cv2.circle(g,(250,300), 180, 255, 25)
cv2.circle(b,(250,300), 180, 0, 25)

cv2.circle(r,(250,300), 155, 20, 25)
cv2.circle(g,(250,300), 155, 125, 25)
cv2.circle(b,(250,300), 155, 17, 25)

cv2.circle(r,(250,300), 130, 0, 25)
cv2.circle(g,(250,300), 130, 255, 25)
cv2.circle(b,(250,300), 130, 255, 25)

cv2.circle(r,(250,300), 105, 30, 25)
cv2.circle(g,(250,300), 105, 20, 25)
cv2.circle(b,(250,300), 105, 255, 25)

cv2.circle(b,(250,300), 80, 128, 25)
cv2.circle(g,(250,300), 80,0 , 25)
cv2.circle(r,(250,300), 80, 128, 25)
```

```python
rainbow=cv2.merge([r,g,b])

plt.imshow(rainbow)
plt.title('Beautiful Rainbow')
plt.axis("off")
plt.show()


```
